# Verilerin Düzenlenmesi ve Birleştirilmesi

```python
import pandas as pd
```

```python
stats = pd.read_csv('Seasons_Stats.csv') #istatistiklerin olduğu data
salary = pd.read_csv("NBA_season1718_salary.csv") #maaşların olduğu data
```

```python
stats.columns
```

```python
#İmzalanan kontratla yani maaşla etkili olabileceğini düşündüğüm, incelenecek spesifik veriler: (+ : pozitif özellik , - : negatif özellik )
#MPG = Minutes Played per Game (Maç başına oynanan süre)(+)
#PPG = Points per Game (Maç başına atılan sayı)(+)
#TOPG = Turnover per Game (Maç başına kaybedilen top)(-)
#PFG = Personal Fouls per Game (Maç başına yapılan bireysel fauller)(-)

#PER = Player Efficiency Rating (Oyuncu verimlilik puanı)(+)
#AGE = Yaş, oyuncu genç ise (+), görece yaşlı ise (-) olduğu düşünüldü

#Veri setinde olmayan spesifik mpg, ppg, topg, pfg verileri hesaplanmalı.

#mpg = mp(minutes played)/g(games played)

#ppg = pts(points)/g(games played)

#topg = tov(turnovers)/g

#pfg = pf(personal fouls)/g
```

```python
#Gerekli hesaplamaları yapıp dataframe'e ekleme
specific_calc = ['MPG','PPG','TOPG','PFG']
calc_from = ['MP','PTS','TOV','PF']

for a,b in zip (specific_calc,calc_from):
    stats[a] = stats[b] / stats['G']
```

```python
#veriler 2017 sezonu sonrasını içermiyor, maaş verileriyle merge etmek için
#stats['Year']>= 2016 ve stats['Year']<2017 almak yeterli
#böylelikle 2016-2017 sezonu istatistikleri oyuncuların bir sonraki sene olan 2017-2018 sezonu maaşlarına 
#nasıl etki etmiş bakabileceğiz
stats.tail(10)
```

```python
#'player' ve 'season17_18' sütunları stats dataframe'i ile merge edilecek 
salary.columns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#stats datasının sadece 2016 ve sonrasını + 2017 öncesini aldık, çünkü salary datası sadece 2017-2018
#sezonunu kapsıyor, bu durumda sadece 2016 sonrasında ve 2017 öncesinde hem salary hem stats datasında bulunan
#oyuncuların verileri merge edilecek
stats = stats[stats['Year']>=2016]
stats = stats[stats['Year']<2017]
#gerekli sütunları alıyoruz
stats = stats[['Player','Year',"Age",'MPG','PPG','PER','TOPG','PFG']]
stats.columns
```

# Merge işlemi ve sonrasında düzenlemeler

```python
#stats ve salary df'ini birleştirme
nba_salary_and_stats = pd.merge(stats , salary)
nba_salary_and_stats.head(10)
```

```python
nba_salary_and_stats.drop_duplicates(subset=['Player'], keep='first',inplace=True)
#çift verilerden kurtulmak için
```

```python
nba_salary_and_stats.drop(columns=['Unnamed: 0', 'Tm'], axis=1, inplace=True)
#gereksiz sütunlardan kurtulma
```

```python
#yıl, maaş sütununu yeniden adlandırma ve maaşı birimi milyona çevirme
nba_salary_and_stats.rename(columns = {'Year':'Stats_from'}, inplace = True)
nba_salary_and_stats.rename(columns = {'season17_18':'2017_2018 Salary'}, inplace = True)
nba_salary_and_stats['2017_2018 Salary'] = nba_salary_and_stats[['2017_2018 Salary']].applymap(lambda line: line/1000000)
```

```python
#merge başarılı ve null değer gözükmüyor 
nba_salary_and_stats.info()
```

```python
nba_salary_and_stats.tail(10)
```

```python
#merge sonrası bozulan indexleri sıfırladık
nba_salary_and_stats.reset_index(inplace=True)
```

```python
nba_salary_and_stats.tail(10)
```

```python
nba_salary_and_stats.drop(['index'],axis=1,inplace=True)
#index sıfırlama sonrası eski index sütununu sildik
```
